# 预测单张图片

```python
# RTMDet --> RTMPose
!/root/miniconda3/envs/myconda/bin/python demo/topdown_demo_with_mmdet.py \
        MMPoseHomework/configs/rtmdet_tiny_ear.py \
        ../mmdetection/work_dirs/rtmdet_tiny_ear/epoch_80.pth \
        MMPoseHomework/configs/rtmpose-s-ear.py \
        work_dirs/rtmpose-s-ear/best_PCK_epoch_290.pth \
        --input mydata/hxsEar.jpg \
        --output-root outputs \
        --device cuda:0 \
        --bbox-thr 0.5 \
        --kpt-thr 0.5 \
        --nms-thr 0.5 \
        --radius 8 \
        --thickness 4 \
        --draw-bbox \
        --draw-heatmap \
        --show-kpt-idx
```
